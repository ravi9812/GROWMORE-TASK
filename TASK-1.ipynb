```python
year=data['iyear'].unique()
years_count=data['iyear'].value_counts(dropna=False).sort_index()
plt.figure(figsize=(15,10))
sns.barplot(x=year,y=years_count,palette='tab10')
plt.xticks(rotation=50)
plt.xlabel('attacking year',fontsize=15)
plt.ylabel('number of attacks a year',fontsize=15)
plt.title('ATTACKS IN YEARS',fontsize=30)
plt.show()

```

here the graph shows number of attacks in each year from 1970 to 1992 here we observed that the number of global terrorism attacks continuosly incresing from 1968.the graph shows 1991 is the most highest attacks takes placed ....

```python
Data=data[['iyear','nkill']].groupby(['iyear']).sum()
fig,ax4=plt.subplots(figsize=(20,10))
Data.plot(kind='bar',alpha=0.7,ax=ax4,color='y')
plt.xticks(rotation=50)
plt.title('people  died due to attack ',fontsize=30)
plt.ylabel('number of killed people',fontsize=20)
plt.xlabel('year',fontsize=20)
top_side=ax4.spines['top']
top_side.set_visible(False)
right_side=ax4.spines['right']
right_side.set_visible(False)
```

this graph shows the number of people killed from year 1970 t0 1992 as the numbe of attacks increases the people killed in attacks also increses.the number of people almost killed in 1964
